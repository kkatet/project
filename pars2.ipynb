```python
import requests, bs4
import re
import json

id_a = 0
data = []

for p in pages:
    s = requests.get(p, timeout = 5)
    b = bs4.BeautifulSoup(s.text, "html.parser")
    links = []
    
    for x in b.find_all("div", attrs = {"class":"img_container big_listing_img"}):
        a1 = r'href="(.+?)"'
        link = re.findall(a1, str(x))
        links += link
    
    for l in links:
        a = requests.get(l)
        c = bs4.BeautifulSoup(a.text, "html.parser")
        
        
        for ww in c.find_all("div", attrs = {"itemprop":"articleBody", "class":"inread-content"}):
            st = r'<(.+?)>'
            article = re.sub(st,'',str(ww))
        
            if len(article) > 500 and len(article) < 2000:
                id_a += 1
                for dd in c.find_all("span", attrs = {"class":"date"}):
                    st = r'<(.+?)>'
                    date_a = re.sub(st,'',str(dd))
                d = dict(id_number = id_a, text = article, district = " ", topic = " ", date = date_a, views = ' ')
                data.append(d)
      
            
```

```python
jsonData = json.dumps(data, sort_keys=False, indent=4, ensure_ascii=False, separators=(',', ': '))

with open("corpus.json", "w", encoding="utf-8") as file:
    file.write(jsonData)
```
